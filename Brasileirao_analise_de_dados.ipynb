### Importando as bibliotecas necessárias

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

### Realizando a raspagem da tabela a ser  tratada e analisada

```python
html= requests.get("https://www.cnnbrasil.com.br/esportes/futebol/tabela-brasileirao-serie-a/").content
#O comando get conecta a bilbioteca requests à URL inserida
#Quando executado com êxito, o código deve retornar [200] como resposta
```

```python
codigo_html = BeautifulSoup(html, 'html.parser') #pega somente o conteúdo em html retirando os espaços
tabela = str(codigo_html.find_all(name = 'table')) #procura somente tabelas dentro do html
df = pd.read_html(tabela)[0] # O índice zero faz com que somente a primeira tabela seja lida
```

```python
display(df)
```

Para facilitar o entendimento, os nomes das colunas serão alterados

```python
df = df.rename(columns={'Classificação': 'Time',
                        'p': 'Total de pontos',
                        'j': 'Número de jogos',
                        'v': 'Vitórias',
                        'e': 'Empates',
                        'd': 'derrotas',
                        'g': 'Gols Pró',
                        'gc': 'Gols contra',
                        'sg': 'Saldo de gols'
                        })
```

Criação de uma nova coluna para registrar a classificação de cada time

```python
df['Classificação'] = range(1, 21)
```

A coluna "Time" contém outros elementos além do nome dos times participantes do campeonato, portanto, é necessário realizar uma limpeza

```python
df['Time'] = df['Time'].str[2:-4]
display(df)
```

Para fins de análise, foi criada uma nova coluna, que relaciona cada time a seu respectivo estado

```python
#Criando um dicionário que relacione o nome de cada time a seu respectivo estado
Estados = {'Red Bull Bragantino': 'SP',
           ' Corinthians': 'SP',
           'Palmeiras' : 'SP',
           ' Santos': 'SP',
           ' São Paulo': 'SP',
          'Botafogo': 'RJ',
           'Flamengo': 'RJ',
           ' Vasco': 'RJ',
           'Fluminense': 'RJ',
           'Atlético-MG': 'MG',
           ' América-MG': 'MG',
           ' Cruzeiro': 'MG',
           'Grêmio': 'RS',
           ' Internacional': 'RS',
           'Fortaleza': 'CE',
           ' Bahia': 'BA',
           'Athletico-PR': 'PR',
           ' Coritiba': 'PR',
           ' Goiás': 'GO',
           ' Cuiabá': 'MT'}
```

```python
#Transformando o estado em uma nova coluna
df['Estado'] = df['Time'].map(Estados)
```

A conexão entre cada time e seu respectivo estado se deu por meio de um dicionário a fim de que a correlação continue correta apesar das atualizações no Dataframe. O mesmo ocorreu com a coluna "Títulos"


```python
#Criando um dicionário que relaciona a quantidade de títulos ao time
Titulos = {'Red Bull Bragantino': 0,
          ' Corinthians': 7,
          'Palmeiras': 11,
          ' Santos': 8,
          ' São Paulo': 6,
          'Botafogo': 2,
          'Flamengo': 7,
          ' Vasco': 4,
          'Fluminense': 4,
          'Atlético-MG': 3,
          ' América-MG': 0,
          ' Cruzeiro': 4,
          'Grêmio': 2,
          ' Internacional': 3,
          'Fortaleza': 0,
          ' Bahia': 2,
          'Athletico-PR': 1,
          ' Coritiba': 1,
          ' Goiás': 0,
          ' Cuiabá': 0}
```

```python
#Transformando o dicionário em uma nova coluna
df['Títulos'] = df['Time'].map(Titulos)
```

```python
# Reorganizando as colunas
df = df[['Classificação', 'Time', 'Estado', 'Total de pontos', 'Número de jogos', 'Vitórias',
             'Empates', 'derrotas', 'Gols Pró', 'Gols contra', 'Saldo de gols', 'Títulos']]
```

```python
display(df)
```

```python
# Salva o DataFrame como um arquivo CSV no ambiente do Google Colab
df.to_csv('tabela_criada.csv', index=False)

#Exibe o conteúdo do diretório atual para verificar o arquivo CSV
import os
os.listdir()
```
